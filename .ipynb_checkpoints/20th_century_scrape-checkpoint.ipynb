```python
# Import libraries

import pandas as pd
import time
import os
import logging
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Key_events_of_the_20th_century/"
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:144.0) Gecko/20100101 Firefox/144.0'}
response = requests.get(url, headers=headers)
```

```python
print(response.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'html.parser')
#print(soup.prettify())  # prints well-formatted HTML
```

```python
text = soup.get_text()

text = text.encode ('utf-8')

with open('20th_century_article_Wiki.txt', 'wb') as f:
       f.write(text)
```
